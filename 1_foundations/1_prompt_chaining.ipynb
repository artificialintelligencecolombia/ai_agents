```python
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
# DANIEL: import and validate Claude's API key

import os
claude_api_key = os.getenv('CLAUDE_API_KEY')

if claude_api_key:
    print(f"OpenAI API Key exists and begins {claude_api_key[:8]}")
else:
    print("OpenAI API Key not set - please head to the troubleshooting guide in the setup folder")
```

```python
# DANIEL: Create client instance
from anthropic import Anthropic

client = Anthropic(
    api_key = os.getenv('CLAUDE_API_KEY')
)
```

```python
prompts_list = [
    "You identify business areas ripe for Agentic AI. Output ONE specific industry/sub-sector with high manual workload and repetitive, data-heavy processes. Respond with: the area name + one sentence on why it's a strong candidate.",
    "Identify ONE challenging, high-cost pain-point in this area that is ripe for an Agentic AI solution above all. Pick something involving multi-step decisions, fragmented data, or expensive human bottlenecks. Respond with: the pain-point + one sentence on why it's hard and costly.",
    "Propose a concrete Agentic AI solution. Cover: (1) what the agent(s) do, (2) the chain/flow of steps, (3) the key outcome/value. Keep it under 150 words, specific and buildable."
]
```

```python
# DANIEL: Agent 1
business_areas_raw = client.messages.create(
    max_tokens=512,
    messages=[{
        "content": "Analyze Colombia micro and macro market. Proceed with 3 candidates",
        "role": "user",
    }],
    model="claude-haiku-4-5-20251001",
    system=prompts_list[0]
)

business_areas = business_areas_raw.content[0].text
print(business_areas)
```

```python
# DANIEL: Agent 2
pain_points_raw = client.messages.create(
    max_tokens=512,
    messages=[{
        "content": business_areas,
        "role": "user",
    }],
    model="claude-haiku-4-5-20251001",
    system=prompts_list[1]
)

pain_points = pain_points_raw.content[0].text
print(pain_points)
```

```python
# DANIEL: Agent 2
agentic_solution_raw = client.messages.create(
    max_tokens=512,
    messages=[{
        "content": pain_points,
        "role": "user",
    }],
    model="claude-haiku-4-5-20251001",
    system=prompts_list[2]
)

agentic_solution = agentic_solution_raw.content[0].text
print(agentic_solution)
```
